#### Exploratory Data Analysis (EDA)


```python
import seaborn as sns

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Creating a series of plots to understand the data better
fig, axes = plt.subplots(3, 2, figsize=(14, 18))

# Plotting distributions of Gender, Education, and Marital Status
sns.countplot(x='sex', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Gender (1=Male, 2=Female)')
axes[0, 0].set_xlabel('Gender')
axes[0, 0].set_ylabel('Count')

sns.countplot(x='education', data=df, ax=axes[0, 1])
axes[0, 1].set_title(
    'Distribution of Education (1=Grad, 2=Univ, 3=High, 4=Others)')
axes[0, 1].set_xlabel('Education')
axes[0, 1].set_ylabel('Count')

sns.countplot(x='marital_status', data=df, ax=axes[1, 0])
axes[1, 0].set_title(
    'Distribution of Marital Status (1=Married, 2=Single, 3=Others)')
axes[1, 0].set_xlabel('Marital Status')
axes[1, 0].set_ylabel('Count')

# Age distribution
sns.histplot(df['age'], bins=30, kde=False, ax=axes[1, 1])
axes[1, 1].set_title('Age Distribution')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Count')

# Default by Gender
sns.countplot(x='sex', hue='default', data=df, ax=axes[2, 0])
axes[2, 0].set_title('Default Status by Gender (1=Male, 2=Female)')
axes[2, 0].set_xlabel('Gender')
axes[2, 0].set_ylabel('Count')

# Default by Education
sns.countplot(x='education', hue='default', data=df, ax=axes[2, 1])
axes[2, 1].set_title('Default Status by Education')
axes[2, 1].set_xlabel('Education')
axes[2, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Now, let's analyze the financial behaviors in the dataset, focusing on the payment history and bill statement amounts.
# For this, we will create a summary of the repayment status across the months and the average bill statement amounts.

# Repayment status across months (X6-X11)
repayment_status = df.loc[:, 'repay_sep':'repay_apr'].apply(
    pd.Series.value_counts, normalize=True)

# Average bill statement amounts across months (X12-X17)
average_bill_amounts = df.loc[:, 'state_sep':'state_apr'].mean()

# Average payment amounts across months (X18-X23)
average_payment_amounts = df.loc[:, 'paid_sep':'paid_apr'].mean()

# Visualizing the repayment status distributions
fig, ax = plt.subplots(figsize=(10, 6))
repayment_status.transpose().plot(kind='bar', stacked=True, ax=ax)
ax.set_title('Normalized Repayment Status from April to September 2005')
ax.set_xlabel('Month')
ax.set_ylabel('Proportion')
ax.legend(title='Repayment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

# Visualizing the average bill and payment amounts
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:red'
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Bill Amount', color=color)
ax1.plot(['Sep', 'Aug', 'Jul', 'Jun', 'May', 'Apr'],
         average_bill_amounts, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
# we already handled the x-label with ax1
ax2.set_ylabel('Average Payment Amount', color=color)
ax2.plot(['Sep', 'Aug', 'Jul', 'Jun', 'May', 'Apr'],
         average_payment_amounts, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Average Bill and Payment Amounts from April to September 2005')
plt.show()

(rep_payment_summary, avg_bill_summary, avg_payment_summary) = (
    repayment_status, average_bill_amounts, average_payment_amounts)
(rep_payment_summary, avg_bill_summary, avg_payment_summary)
```
