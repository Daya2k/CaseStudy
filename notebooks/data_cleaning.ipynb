```python
# Now, let's analyze the financial behaviors in the dataset, focusing on the payment history and bill statement amounts.
# For this, we will create a summary of the repayment status across the months and the average bill statement amounts.

# Repayment status across months (X6-X11)
repayment_status = df.loc[:, 'repay_sep':'repay_apr'].apply(
    pd.Series.value_counts, normalize=True)

# Average bill statement amounts across months (X12-X17)
average_bill_amounts = df.loc[:, 'state_sep':'state_apr'].mean()

# Average payment amounts across months (X18-X23)
average_payment_amounts = df.loc[:, 'paid_sep':'paid_apr'].mean()

# Visualizing the repayment status distributions
fig, ax = plt.subplots(figsize=(10, 6))
repayment_status.transpose().plot(kind='bar', stacked=True, ax=ax)
ax.set_title('Normalized Repayment Status from April to September 2005')
ax.set_xlabel('Month')
ax.set_ylabel('Proportion')
ax.legend(title='Repayment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

# Visualizing the average bill and payment amounts
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:red'
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Bill Amount', color=color)
ax1.plot(['Sep', 'Aug', 'Jul', 'Jun', 'May', 'Apr'],
         average_bill_amounts, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
# we already handled the x-label with ax1
ax2.set_ylabel('Average Payment Amount', color=color)
ax2.plot(['Sep', 'Aug', 'Jul', 'Jun', 'May', 'Apr'],
         average_payment_amounts, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Average Bill and Payment Amounts from April to September 2005')
plt.show()

(rep_payment_summary, avg_bill_summary, avg_payment_summary) = (
    repayment_status, average_bill_amounts, average_payment_amounts)
(rep_payment_summary, avg_bill_summary, avg_payment_summary)
```

#### Feature Engineering


```python
for col in df.columns:
    Q1 = df[col].quantile(.25)
    Q3 = df[col].quantile(.75)
    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    print(f"{col} ==> \tupper_limit : {upper_limit} , lower_limit : {lower_limit}")
```

```python
def outlier(df):
    columns = ['credit_amt', 'state_sep', 'state_aug', 'state_jul', 'state_jun', 'state_may',
               'state_apr', 'paid_sep', 'paid_aug', 'paid_jul', 'paid_jun', 'paid_may',
               'paid_apr']
    for col in columns:
        mean = df[col].mean()
        std = df[col].std()
        upper_limit = mean + 4 * std
        lower_limit = mean - 4 * std
        print(f"{col} ==> \tupper_limit : {upper_limit} , lower_limit : {lower_limit}")
        df = df[(df[col] > lower_limit) & (df[col] < upper_limit)]
    return df
```

```python
new_df = pd.concat([x_sm, y_sm], axis=1)
```
